# Charge Report Extraction Script Explanation

This script, `charge_reports_to_info_csv_llm.py`, converts legal charge report markdown files into a structured CSV file for data analytics.

## 🔧 What it does

1. Reads markdown charge report files from `ChargeReports/`.
2. Sends each file text to an LLM (LiteLLM/OpenAI chat completion API) with a fixed extraction prompt.
3. Parses the returned JSON and maps fields into a CSV row format.
4. Writes an output CSV file with columns required for "Info to Extract".

## 🧠 Main components

- **LLM prompt**: The script uses `LLM_PROMPT` to ask the model to extract specific fields (case, charge, accused/victim details, offence group, special type) in strict JSON.
- **`call_llm(text, client)`**: Calls the OpenAI client, extracts JSON from markdown code fences, then loads it into a Python dictionary.
- **`to_csv_rows(extracted)`**: Converts extraction results into one or two CSV rows (accused and victim if available), with fields.
- **`main()`**: Parses command-line args, loads charge report files, calls the LLM per file, handles errors, then writes CSV.

## 📝 Input and output

- Input: Markdown files in `ChargeReports/` (or a single file via `--charge`), each describing a charge report.
- Output: CSV file at `--output` (default: `Info to Extract Test.csv`) with fields:
  - `Case`, `Charge`, `Entity`, `Entity Type`, `Age`, `Gender`, `Relationship to Victim`, `Offence Groups`, `Special Type`

## 🚀 Usage

From the script directory:

```bash
python charge_reports_to_info_csv_llm.py
python charge_reports_to_info_csv_llm.py --sample 50
python charge_reports_to_info_csv_llm.py --charge DAC-000001-2023
```

## ⚠️ Notes

- The script expects `openai` package installed and environment variables for API key (`LLM_API_KEY`) and base URL (`LLM_BASE_URL`) configured.
- It uses a fallback model `agc-llmaas-claude-4-5-dev` and default key placeholder.
- If the LLM output is not valid JSON or extraction fails, the script logs errors and continues.
